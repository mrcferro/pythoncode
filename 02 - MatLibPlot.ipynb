# Bibliotecas de Visualização: MatLibPlot e CV2 
MatLibPlot - https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
# criando e imprimindo arrays

salario = [1100.00, 1045.00, 1039.00, 998.00, 954.00, 937.00, 880.00, 788.00, 724.00, 678.00, 622.00]
ano     = [   2021,    2020,    2019,   2018,    2017,  2016,   2015,   2014,   2013,   2012,   2011]

print(salario)
print(ano)
```

```python
# plotando um grafico com MATLIBPLOT (Biblioteca de Visualização)

import matplotlib.pyplot as plt
plt.plot(salario,ano)
plt.show
```

```python
# melhorando o gráfico anterior...

plt.xlabel ("Valor em R$")
plt.ylabel ("Anos")
plt.grid()
plt.plot(salario,ano, linestyle='--', marker='s')
plt.show
```

```python
# mais um exemplo...

dados = {'Cap1':10, 'Cap2':20, 'Cap3':20, 'Cap4':18, 'Cap5':2, 'Cap6':5, 'Cap7':2, 'Cap8':8, 'Cap9':15 }
caps = list(dados.keys())
questoes = list(dados.values())

print(caps)
print(questoes)
```

```python
fig, ax = plt.subplots()
ax.bar(caps, questoes,align='center')
```

```python
fig, ax = plt.subplots()
ax.barh(caps, questoes,align='center')
ax.set_ylabel('Capítulos')
ax.set_title('Capítulos vs. Porcentagem de Questões')
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 3), sharey=True)
axs[0].bar(caps, questoes)
axs[1].scatter(caps, questoes)
axs[2].plot(caps, questoes)
fig.suptitle('Capítulos vs. % Questões')
```
